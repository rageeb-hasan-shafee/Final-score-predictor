```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./dataset/final_ball_by_ball_first_innings.csv')
```

```python
players = np.union1d(df['batter'].unique(), df['non_striker'].unique())
print(f'batters: {players.shape[0]}')
players = np.union1d(players, df['bowler'].unique())
print(f'players: {players.shape[0]}')
```

```python
players = set(df['non_striker'].unique()) - set(df['batter'].unique()) 
print(f'only batters: {len(players)}')
```
